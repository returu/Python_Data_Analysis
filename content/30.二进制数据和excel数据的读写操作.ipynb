# 2.excel数据读写：

### 2.1 读取excel文件

#### ExcelFile类

```python
# 使用 pandas.ExcelFile 类来创建一个实例
xlsx = pd.ExcelFile("./data/PM25.xlsx")
```

```python
# 该对象可以显示文件中可用工作表名称的列表
xlsx.sheet_names
```

```python
# 然后使用 parse 将存储在工作表中的数据读入 DataFrame 中
xlsx.parse(sheet_names="Sheet1")
```

```python
# 此时 Excel 表有一个索引列，可以通过 index_col 参数取消
xlsx.parse(sheet_name="Sheet1", index_col=0)
```

#### read_excel()

```python
xlsx = pd.read_excel("./data/PM25.xlsx",sheet_name="Sheet1", index_col=0)
xlsx
```

### 2.2 写入excel文件

#### ExcelWriter类

```python
writer = pd.ExcelWriter("./data/PM25.xlsx")
df.to_excel(writer,"Sheet1")
writer.save()
```

#### to_excel()

```python
df.to_excel('./data/PM25.xlsx')
```
