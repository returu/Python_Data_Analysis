```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'key': ['x', 'y', 'z', 'x', 'y', 'z', 'x', 'y', 'z', 'x', 'y', 'z'],
                   'value1': np.arange(12), 'value2': np.arange(12,24)},
                  index=['a','b','c','d','e','f','g','h','i','j','k','l'])
df
```

### 按 key列 进行分组

```python
grouped = df.groupby('key')
```

```python
for name , value in grouped:
    print(name)
    print(value)
```

### 根据分组计算平均值

```python
grouped.agg("mean")
```

```python
grouped.apply("mean")
```

```python
grouped.transform("mean")
```

### 使用自定义函数将每个组乘以 2

```python
def time_two(group):
    return group * 2
```

```python
grouped.apply(time_two)
```

```python
grouped.transform(time_two)
```

### apply 允许对dataframe的多个series进行操作，而 transform 只能操作一列

```python
def sum_func(group):
    return (group["value1"] + group["value2"])
```

```python
df.groupby('key').apply(sum_func)
```

```python
df.groupby('key').transform(sum_func)
```
