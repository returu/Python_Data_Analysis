```python
import numpy as np
import pandas as pd
```

# 1.数据库风格的DataFrame连接：

```python
df1 = pd.DataFrame({'user_id':[1,2,3,4,5,6,7],'device':['Iphone','Huawei','HTC','oppo','vivo','xiaomi','google']})
df1
```

```python
df2 = pd.DataFrame({'user_id':[1,2,3,4,5,6,8],'month_G':[5,9,8,10,15,30,2]})
df2
```

```python
pd.merge(df1,df2,on='user_id')
```

#### 分别指定连接键

```python
df1 = pd.DataFrame({'user_id':[1,2,3,4,5,6,7],'device':['Iphone7','Huawei','HTC','oppo','vivo','xiaomi','google']})
df1
```

```python
df2 = pd.DataFrame({'id':[1,2,3,4,5,6,8],'month_G':[5,9,8,10,15,30,2]})
df2
```

```python
pd.merge(df1,df2,left_on='user_id',right_on='id')
```

#### 连接模式

```python
df1 = pd.DataFrame({'user_id':[1,2,3,4,5,6,7],'device':['Iphone7','Huawei','HTC','oppo','vivo','xiaomi','google']})
df1
```

```python
df2 = pd.DataFrame({'user_id':[1,2,3,4,5,6,8],'month_G':[5,9,8,10,15,30,2]})
df2
```

```python
pd.merge(df1,df2,how='outer')
```

```python
pd.merge(df1,df2,how='left')
```

#### 使用多个键进行连接

```python
df1 = pd.DataFrame({'user_id':[1,2,3,4,5,6,7],'device':['Iphone','Huawei','HTC','oppo','vivo','xiaomi','google'],'key':['shandong','shanghai','beijing','hebei','jiangsu','hunan','henan']})
df1
```

```python
df2 = pd.DataFrame({'user_id':[1,2,3,4,5,6,8],'month_G':[5,9,8,10,15,30,2],'key':['shandong','shanghai','beijing','hebei','jiangsu','hunan','jiangxi']})
df2
```

```python
pd.merge(df1,df2,on=['user_id','key'],how='outer')
```

#### 重叠列名的处理

```python
df1 = pd.DataFrame({'user_id':[1,2,3,4,5,6,7],'device':['Iphone','Huawei','HTC','oppo','vivo','xiaomi','google'],'key':['shandong','shanghai','beijing','hebei','jiangsu','hunan','henan']})
df1
```

```python
df2 = pd.DataFrame({'user_id':[1,2,3,4,5,6,8],'month_G':[5,9,8,10,15,30,2],'key':['ios','shanghai','beijing','hebei','jiangsu','hunan','jiangxi']})
df2
```

```python
pd.merge(df1,df2,on=['user_id'],how='outer')
```

```python
pd.merge(df1,df2,on=['user_id'],how='outer',suffixes=["_df1","_df2"])
```
