```python
import pandas as pd
```

# 3.read_csv()：

```python
df = pd.read_csv("data/PM25.csv",encoding='gb18030')
df
```

#### names参数

```python
# 读取无表头文件
df = pd.read_csv("data/PM25_no_header.csv",encoding='gb18030',header=None)
df
```

```python
# 指定自定义列名
names = ["city" , "day" , "months" , "province" , "region"]

df = pd.read_csv("data/PM25_no_header.csv",encoding='gb18030',header=None,names=names)
df
```

### index_col 参数

```python
# 使用“城市”列作为索引
df = pd.read_csv("data/PM25.csv",encoding='gb18030' , index_col="城市")
df
```

```python
# 使用“省份”、所在区域”两列组成分层索引
df = pd.read_csv("data/PM25.csv",encoding='gb18030' , index_col=["所在区域","省份"])
df
```

### sep参数

```python
# 将分隔符设置为空格
df = pd.read_csv("data/PM25_sep.csv",encoding='gb18030' , sep=" ")
df
```

```python
# 将一个正则表达式传递给分隔符参数
df = pd.read_csv("data/PM25_sep.csv",encoding='gb18030' , sep="\s+")
df
```

### skiprows 参数

```python
# 跳过第二行和第九行
df = pd.read_csv("data/PM25.csv",encoding='gb18030' , skiprows=[1,8])
df
```

### 缺失值处理

```python
# 读取有缺失值的文件
df = pd.read_csv("data/PM25_nan.csv",encoding='gb18030')
df
```

```python
# 查看缺失值情况
pd.isna(df)
```

```python
# keep_default_na
df = pd.read_csv("data/PM25_nan.csv",encoding='gb18030',keep_default_na=False)
df
```

```python
pd.isna(df)
```

```python
# na_values
df = pd.read_csv("data/PM25_nan.csv",encoding='gb18030' , na_values=0)
df
```

```python
sentinels = {"超标天数":[0] , "超标月数":[0,999]}

df = pd.read_csv("data/PM25_nan.csv",encoding='gb18030' , na_values=sentinels)
df
```
