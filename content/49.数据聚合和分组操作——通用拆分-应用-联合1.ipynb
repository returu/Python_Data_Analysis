```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("./data/happiness_report.csv")
```

```python
df
```

### 假设你想按Year作为分组选出 Happiness Score 值最高的五个。

```python
def top(df,n=5,column="Happiness_Score"):
    return df.sort_values(column , ascending=False)[:n]
```

```python
top(df , n=4)
```

```python
df.groupby("Year").apply(top)
```

### 除了可以向 apply 传递函数，也可以传递其他参数或关键字

```python
df.groupby("Year").apply(top , n=2 , column="Economy")
```

# 1.压缩分组键：

```python
df.groupby("Year" ,group_keys=False).apply(top, n=3 )
```

```python
df.groupby("Year" ,as_index=False).apply(top, n=3 )
```

# 2.分位数和桶分析：

```python
df_2015 = df[df["Year"]==2015]
df_2015
```

### pandas.cut 和 pandas.qcut

```python
quartiles = pd.cut(df_2015["Economy"]  , 4)
# quartiles = pd.cut(df_2015["Economy"]  , 4 , labels=["Q1","Q2","Q3","Q4"])
```

```python
quartiles
```

### cut返回的Categorical对象可以直接传递给groupby，所以我们可以计算出一组四分位数的分组统计数据

```python
def get_stats(group):
    return pd.DataFrame({
        "min": group.min(),
        "max": group.max(),
        "count": group.count(),
        "mean": group.mean()})
```

```python
grouped = df_2015.groupby(quartiles)
```

```python
grouped[["Happiness_Score"]].apply(get_stats)
```

### 可以通过 agg 更简单地计算相同的结果

```python
grouped[["Happiness_Score"]].agg(["min", "max", "count", "mean"])
```

### pandas.qcut

```python
# quartiles_samp = pd.qcut(df_2015["Economy"] , 4)
quartiles_samp = pd.qcut(df_2015["Economy"] , 4, labels=False)
```

```python
quartiles_samp
```

```python
grouped = df_2015.groupby(quartiles_samp)
```

```python
grouped[["Happiness_Score"]].apply(get_stats)
```
